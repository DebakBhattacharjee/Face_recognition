```python
from sklearn.datasets import fetch_olivetti_faces
import pandas as pd
import numpy as np
```

```python
olivetti_faces = fetch_olivetti_faces()
olivetti_faces
```

```python
len(olivetti_faces.data)  # 400 images in dataset
```

```python
len(olivetti_faces.data[0])    # Each row corresponds to a ravelled face image of original size 64 x 64 pixels.
```

```python
target = olivetti_faces.target
target
```

```python
features = olivetti_faces.data
features
```

```python
import matplotlib.pyplot as plt
```

# Showing 1 image from 40 people

```python
fig, sub_plots = plt.subplots(nrows = 5, ncols = 8, figsize = (14, 8))
sub_plots = sub_plots.flatten()    #making the 2d array of the subplot into 1d for better iteration

for unique_user_id in np.unique(target):
    image_index = unique_user_id * 8
    sub_plots[unique_user_id].imshow(features[image_index].reshape(64, 64), cmap = 'gray')
    sub_plots[unique_user_id].set_xticks([])
    sub_plots[unique_user_id].set_yticks([])
#    sub_plots[unique_user_id].set_title("Face id: %s", unique_user_id)
    
plt.suptitle('Face Id (40 people)')
plt.show()
```

```python
fig, sub_plots = plt.subplots(nrows = 1, ncols = 10, figsize = (14, 8))
for j in range(10):
    sub_plots[j].imshow(features[j].reshape(64, 64), cmap = 'gray')
    sub_plots[j].set_xticks([])
    sub_plots[j].set_yticks([])
    
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, stratify=target, random_state=0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=100, whiten=True)
pca.fit(features)
X_pca = pca.fit_transform(features)
```

```python
plt.figure(1, figsize=(12, 8))
plt.plot(pca.explained_variance_, linewidth = 2)
plt.xlabel('Components')
plt.ylabel('Explained variances')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
models = [("Logistic Regression", LogisticRegression()), ("Support Vector Machine", SVC()), ("Naives Bayes", GaussianNB())]
```

```python
for name, model in models:
    
    kfold = KFold(n_splits=5, shuffle= True, random_state=0)
    cross_val = cross_val_score(model, X_pca, target, cv=kfold)
    print("Mean of cross validation of %s" %name, " : %s" %cross_val.mean())
```
